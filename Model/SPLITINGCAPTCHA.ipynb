```python
from PIL import Image
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import cv2
import imutils

# Set the path to the folder containing the captcha images
folder_path = "./captcha-dataset-master/captcha-dataset-master/"

# Load the labels from the text file into a pandas dataframe
labels_df = pd.read_csv("./captcha-dataset-master/captcha-dataset-master/yzm_labels.txt", sep=",", header=None, names=["image_path", "label"])
```

```python
OUTPUT_DIR = "outputs"  # Replace with the directory to save the split character images

counts = {}

for index, row in labels_df.iterrows():
    image_path = row["image_path"]
    captcha_label = row["label"]
    #cv2.imshow(image_path, imutils.resize(cv2.imread(os.path.join(folder_path, image_path)), width=28))
    print(captcha_label)

    try:
        # Load the image using OpenCV
        img = cv2.imread(os.path.join(folder_path, image_path))
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        gray = cv2.copyMakeBorder(gray, 8, 8, 8, 8, cv2.BORDER_REPLICATE)
        thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY_INV | cv2.THRESH_OTSU)[1]
        cnts = cv2.findContours(thresh.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
        cnts = cnts[0]
        cnts = sorted(cnts, key=lambda c: cv2.boundingRect(c)[0])[:4]

        for i, c in enumerate(cnts):
            (x, y, w, h) = cv2.boundingRect(c)

            if w > 40:
                roi = gray[y-5:y+5+h, x-5:x + int(w/2)]
                roi_2 = gray[y-5:y+5+h, x + int(w/2):x+5+int(w/2)]
                rois = [roi, roi_2]
            else:
                roi = gray[y-5:y+5+h, x-5:x+5+w]
                rois = [roi]

            if isinstance(captcha_label, str):
                label = captcha_label[i] if i < len(captcha_label) else "UNKNOWN"
            else:
                label = "UNKNOWN"

            for i, roi in enumerate(rois):
                print(f"Saving ROI {i+1} with label '{label}'")
                if roi is not None and roi.size != 0:
                    dir_path = os.path.join(OUTPUT_DIR, label)
                    if not os.path.exists(dir_path):
                        os.makedirs(dir_path)
                    count = counts.get(label, 1)
                    p = os.path.join(dir_path, "{}.png".format(str(count).zfill(6)))
                    cv2.imwrite(p, roi)
                    counts[label] = count + 1

    except FileNotFoundError:
        print(f"Image not found for index {index}. Skipping...")
```

```python
# Get a list of all subfolders (labels) in the output directory
labels = os.listdir(OUTPUT_DIR)

# Randomly pick 4 labels
selected_labels = np.random.choice(labels, size=4, replace=False)

# Load and concatenate the selected character images
captcha_image = np.zeros((48, 120, 1), dtype=np.uint8)

for i, label in enumerate(selected_labels):
    # Get the images for the current label
    label_dir = os.path.join(OUTPUT_DIR, label)
    image_files = os.listdir(label_dir)

    # Randomly pick one image from the label folder
    image_file = np.random.choice(image_files)

    # Load the image and resize to (48, 30)
    image_path = os.path.join(label_dir, image_file)
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    image = cv2.resize(image, (30, 48), interpolation=cv2.INTER_AREA)

    # Add the character image to the captcha image
    captcha_image[:, i * 30:(i + 1) * 30] = image[:, :, np.newaxis]

# Display the complete captcha image
cv2.imshow("Complete Captcha", captcha_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
